```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['month_number'],df['total_profit'],linestyle='dotted',marker='o')
plt.legend('total profit',loc="upper left", fontsize=12)
plt.title('Total profit based on month')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.show()
```

```python
march=df[df['month_number']==3]
march
```

```python
products=march.columns.tolist()[1:7]
sales=march.values[0][1:7].tolist()
exp=[0,0,0.2,0,0,0]
plt.pie(sales,labels=products,explode=exp,shadow=True,autopct='%0.1f%%',colors=['skyblue','lightgreen','pink','green','yellow','orange'])
plt.title('Sales in March',loc='right')
plt.show()
```

```python
quaters=['Q1','Q2','Q3','Q4']
quaterly_facecream=[sum(df['facecream'][0:3]),sum(df['facecream'][3:6]),sum(df['facecream'][6:9]),sum(df['facecream'][9:12])]
quaterly_facewash=[sum(df['facewash'][0:3]),sum(df['facewash'][3:6]),sum(df['facewash'][6:9]),sum(df['facewash'][9:12])]
quaterly_toothpaste=[sum(df['toothpaste'][0:3]),sum(df['toothpaste'][3:6]),sum(df['toothpaste'][6:9]),sum(df['toothpaste'][9:12])]
quaterly_bathingsoap=[sum(df['bathingsoap'][0:3]),sum(df['bathingsoap'][3:6]),sum(df['bathingsoap'][6:9]),sum(df['bathingsoap'][9:12])]
quaterly_shampoo=[sum(df['shampoo'][0:3]),sum(df['shampoo'][3:6]),sum(df['shampoo'][6:9]),sum(df['shampoo'][9:12])]
quaterly_moisturizer=[sum(df['moisturizer'][0:3]),sum(df['moisturizer'][3:6]),sum(df['moisturizer'][6:9]),sum(df['moisturizer'][9:12])]
```

```python
bar_width = 0.15
index = np.arange(len(quaters))

# Plot
plt.figure(figsize=(12,6))

# Plot bars for each product, with slight offset for each product
plt.bar(index, quaterly_facecream, bar_width, label='Facecream', color='pink')
plt.bar(index + bar_width, quaterly_facewash, bar_width, label='Facewash', color='green')
plt.bar(index + 2 * bar_width, quaterly_toothpaste, bar_width, label='Toothpaste', color='skyblue')
plt.bar(index + 3 * bar_width, quaterly_bathingsoap, bar_width, label='Bathingsoap', color='grey')
plt.bar(index + 4 * bar_width, quaterly_shampoo, bar_width, label='Shampoo', color='yellow')
plt.bar(index + 5 * bar_width, quaterly_moisturizer, bar_width, label='Moisturizer', color='orange')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Quarter-wise Sales Data for All Products')
plt.legend(loc='upper right')
plt.xticks(index + 2.5 * bar_width, quaters)
plt.show()
```
