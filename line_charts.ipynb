# plt.title() , xlabel() , ylabel()

```python
plt.figure(figsize=(12,5))

plt.plot(df['index'],df['RG Sharma'],marker='D')
plt.plot(df['index'],df['V Kohli'],marker='<')

plt.title(" V kohli and Rohit Sharama carrer comparision ")
plt.xlabel("Seasons")
plt.ylabel("runs")


plt.show()
```

# color

```python
plt.figure(figsize=(12,5))

plt.plot(df['index'],df['RG Sharma'],marker='D',color='green')
plt.plot(df['index'],df['V Kohli'],marker='<',color='red')

plt.title(" V kohli and Rohit Sharama carrer comparision ")
plt.xlabel("Seasons")
plt.ylabel("runs")


plt.show()
```

# legend

```python
plt.figure(figsize=(12,5))

plt.plot(df['index'],df['RG Sharma'],marker='D',color='green',label='Rohit')
plt.plot(df['index'],df['V Kohli'],marker='<',color='red',label='kohli')

plt.title(" V kohli and Rohit Sharama carrer comparision ")
plt.xlabel("Seasons")
plt.ylabel("runs")

#plt.legend()
#plt.legend(loc=('upper left'))
#plt.legend(loc=('centexzxzr'))
plt.legend(loc='best')

plt.show()
```

# linestyle 

```python
plt.figure(figsize=(12,5))

plt.plot(df['index'],df['RG Sharma'],marker='D',color='green',label='Rohit',linewidth=2,linestyle='--')
plt.plot(df['index'],df['V Kohli'],marker='<',color='red',label='kohli',linewidth=2,linestyle='dotted')

plt.title(" V kohli and Rohit Sharama carrer comparision ")
plt.xlabel("Seasons")
plt.ylabel("runs")

#plt.legend()
#plt.legend(loc=('upper left'))
#plt.legend(loc=('centexzxzr'))
plt.legend(loc='best')

plt.show()
```

# linewidth 

```python
plt.figure(figsize=(12,5))

plt.plot(df['index'],df['RG Sharma'],marker='D',color='orange',label='Rohit',linewidth=3,markersize=10)
plt.plot(df['index'],df['V Kohli'],marker='p',color='red',label='kohli',linewidth=3,markersize=7)

plt.title(" V kohli and Rohit Sharama carrer comparision ")
plt.xlabel("Seasons")
plt.ylabel("runs")

#plt.legend()
#plt.legend(loc=('upper left'))
#plt.legend(loc=('centexzxzr'))
plt.legend(loc='best')

plt.show()
```
