```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Create the bar plot
plt.figure(figsize=(10,3))
bars = plt.bar(categories, values)

# Add the value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, str(yval), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 15, 7, 12]

# Custom x-axis scale data
x_ticks = [10,20,30,40]  # Custom positions for the x-axis ticks
x_labels = ['A', 'B', 'C', 'D']  # Custom labels for each category

# Create the bar plot
plt.bar(x_ticks, values)


# Set custom x-axis labels and scale
plt.xticks(ticks=x_ticks, labels=x_labels)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 15, 7, 12]

x_ticks = [10,11,12,13]  # Custom positions for the x-axis ticks
x_labels = ['A', 'B', 'C', 'D']  # Custom labels for each category

bars=plt.bar(x_ticks, values)
plt.xticks(ticks=x_ticks, labels=x_labels)

for i in range(len(bars)):
    plt.text(x_ticks[i],bars[i].get_height(),values[i])
   

plt.show()
```

```python
# simple data
data = [23,45,100,20,49]

subjects = ['eng','science','maths','sst','nepali']

explode=[0.01,0.2,0,0,0.3]

#plt.pie(data,autopct='%0.1f%%',colors=colors,shadow=True,explode=explode)

plt.pie(data,labels=subjects,autopct='%0.1f%%',shadow=True,explode=explode,colors=['blue','green','yellow','pink','cyan'])
plt.show()
```

```python
marks_sub=[]
for i in range(len(data)):
    sub=subjects[i]
    mark=str(data[i])
    det=sub+' : '+mark
    marks_sub.append(det)
    
```

```python
plt.pie(data,labels=marks_sub,autopct='%0.1f%%',shadow=True,explode=explode,colors=['blue','green','yellow','pink','cyan'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 15, 7, 12]

# Custom x-axis scale: Numbers on the x-axis instead of category names
x_ticks = np.arange(1, len(categories) + 1)  # Numeric x-axis scale
x_labels = [1, 2, 3, 4]  # Numeric labels corresponding to the ticks

# Create the bar plot
plt.bar(x_ticks, values)

# Set custom x-axis labels
plt.xticks(ticks=x_ticks, labels=x_labels)

# Show the plot
plt.show()
```
