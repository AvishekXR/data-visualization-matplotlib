```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('polluted.csv')
df
```

```python
df.columns
```

```python
x=df[df['Country'].isin(['Iran','China'])]
data=x.groupby('Year')['PM2.5'].sum()
year=data.index.tolist()
PM2=data.values
plt.plot(year,PM2,marker='o', markerfacecolor='yellow',label='PM2.5')
plt.xticks(year)
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.legend()
plt.title('Relation between year and sum of PM2.5 of China and Iran')
plt.show()
```

```python
plt.hist(df['PM10'],bins=12,density=True,alpha=0.7,color='pink', edgecolor='black',label='PM10')
plt.xlabel('PM10')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
x=df[df['Country'].isin(['Poland','Chile'])]
plt.scatter(x['PM2.5'],x['PM10'],marker='+',color='orange')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.title('Relation between PM10 and PM2.5 of Poland and Chile')

plt.show()
```

```python
country=df['Country'].value_counts().nlargest(5).index
freq=df['Country'].value_counts().nlargest(5).values
country
```

```python
explode=[0,0.2,0,0,0]
plt.pie(freq,labels=country,autopct='%0.1f%%',explode=explode,shadow=True)
plt.title('Top 5 most frequent countries')
plt.show()
```
